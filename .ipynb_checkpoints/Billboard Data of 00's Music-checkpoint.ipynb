## Step 2: Clean your data.

##### Do some rudimentary cleaning. Rename any columns that are poorly named, shorten any strings that may be too long, check for missing values (and replace them if it makes sense to do so). Explain your rationale for the way you choose to "impute" the missing data.

```python
# Rename columns by shortening the names with the CamelCase naming convention
data.columns.values
data.rename(columns = {'artist.inverted':'artist', 'date.entered': 'enterDate', 'date.peaked':'peakedDate'}, inplace = True)
```

```python
# Check the datatype of each column, change the the "x1st.week" column from int to float
data.dtypes
data['x1st.week'] = data['x1st.week'].astype(float)

# Add a column that calculates the length of a song in seconds
def getSeconds(timgString):
    x = dt.datetime.strptime(timgString, "%M:%S")
    return datetime.timedelta(hours=x.hour,minutes=x.minute,seconds=x.second).total_seconds()

data['timeInSeconds'] = pd.Series([getSeconds(x) for x in data['time']], index = data.index)
```

```python
# Add a column as a unique identifier of a track
data['artistTrack'] = data['artist'] + " - '" + data['track'] + "'"
```

```python
# Change the enterDate and peakedDate columns from string to datetime format
data['enterDate'] = pd.Series([datetime.datetime.strptime(x,'%Y-%m-%d') for x in data['enterDate']], index = data.index)
data['peakedDate'] = pd.Series([datetime.datetime.strptime(x,'%Y-%m-%d') for x in data['peakedDate']], index = data.index)

# Add a column that calculates the days it took a song to reach peak popularity
data['weeksBeforePeak'] = (data['peakedDate'] - data['enterDate'] ).astype('timedelta64[W]')

# Make the genre column categorical
data['genre'] = data['genre'].astype("category")
```

```python
# Reorder the columns
columnNames = data.columns.values.tolist()
columnNames = columnNames[:3] + [columnNames[-2]] + [columnNames[3]] + [columnNames[-3]] + columnNames[4:7] + [columnNames[-1]] + columnNames[7:-3]  
data = data[columnNames]
```

##### Using Pandas' built in `melt` function, pivot the weekly ranking data to be long rather than wide. As a result, you will have removed the 72 'week' columns and replace it with two: Week and Ranking. There will now be multiple entries for each song, one for each week on the Billboard rankings.

```python
# Pivot weekly ranking data to the long format using the melt function
melted_data = pd.melt(data, id_vars = list(data.ix[:, :'weeksBeforePeak'].columns.values), var_name = "week", value_name = 'rank')
```

```python
melted_data.head(4)
```

```python
# Convert the 'week' column from string type to float
def getWeeks(weekString):
    weekName = x.split('.')[0]
    return float(weekName[1:len(weekName)-2]) 
    
melted_data['week'] = pd.Series([getWeeks(x) for x in melted_data['week']], index = melted_data.index)
```

```python
# Calculate the peak ranking, life span (numbers of weeks in Top 100) and number of weeks at peak for each track
trackPeakRank = pd.pivot_table(melted_data[~melted_data['rank'].isnull()], index = ['artistTrack'], values = ['rank'], aggfunc = [np.min])['amin']
trackLifeSpan = pd.pivot_table(melted_data[~melted_data['rank'].isnull()], index = ['artistTrack'], values = ['rank'], aggfunc = [len])['len']
trackPeakWeeks = pd.pivot_table(melted_data[~melted_data['rank'].isnull()], index = ['artistTrack'], values = ['rank'], aggfunc = lambda x: len(x[x == np.min(x)]))

trackPeakRank.reset_index(level=0, inplace=True)
trackLifeSpan.reset_index(level=0, inplace=True)
trackPeakWeeks.reset_index(level=0, inplace=True)

trackPeakRank.rename(columns={'rank':'peakRank'}, inplace = True)
trackLifeSpan.rename(columns={'rank':'lifeSpan'}, inplace = True)
trackPeakWeeks.rename(columns={'rank':'weeksAtPeak'}, inplace = True)

# Add the peak ranking, life span and weeks at peak columns to the melted_data Data Frame
melted_data = pd.merge(melted_data, trackPeakRank, how='inner', on='artistTrack')
melted_data = pd.merge(melted_data, trackLifeSpan, how='inner', on='artistTrack')
melted_data = pd.merge(melted_data, trackPeakWeeks, how='inner', on='artistTrack')

melted_data['weeksAfterPeak'] = melted_data['lifeSpan'] - melted_data['weeksBeforePeak'] - melted_data['weeksAtPeak']
```

```python
# Reorder the columns of melted_data
columnNames = melted_data.columns.values.tolist()
columnNames = columnNames[:10] + columnNames[-2:] + [columnNames[-3]] + [columnNames[-4]] + columnNames[-6:-4]
melted_data = melted_data[columnNames]
```

```python
# Create a new Data Frame to store the Track level info
trackData = melted_data.ix[:, (melted_data.columns !='week') & (melted_data.columns != 'rank')].drop_duplicates()
```
